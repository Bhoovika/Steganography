#Edge detection-sobel

```python
import torch
import torch.nn as nn
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

class SobelEdgeDetector(nn.Module):
    def __init__(self):
        super(SobelEdgeDetector, self).__init__()
        self.conv_x = nn.Conv2d(1, 1, kernel_size=3, stride=1, padding=1, bias=False)
        self.conv_y = nn.Conv2d(1, 1, kernel_size=3, stride=1, padding=1, bias=False)

        #vertical filter
        sobel_x = torch.tensor([[-1, 0, 1],
                                [-2, 0, 2],
                                [-1, 0, 1]], dtype=torch.float32)
        # horizontal filter
        sobel_y = torch.tensor([[-1, -2, -1],
                                [0, 0, 0],
                                [1, 2, 1]], dtype=torch.float32)

        self.conv_x.weight = nn.Parameter(sobel_x.unsqueeze(0).unsqueeze(0))
        self.conv_y.weight = nn.Parameter(sobel_y.unsqueeze(0).unsqueeze(0))

    def forward(self, x):
        x_gradient_x = self.conv_x(x)
        x_gradient_y = self.conv_y(x)
        x_gradient = torch.sqrt(x_gradient_x ** 2 + x_gradient_y ** 2)
        return x_gradient

model = SobelEdgeDetector()

image_path = '/content/drive/MyDrive/image/image_rgb_1/preprocessed_image_rgb.jpg'
image = Image.open(image_path).convert('L')
image = image.resize((1500, 1500)) 
image = torch.tensor(np.array(image)).unsqueeze(0).unsqueeze(0).float() 

edge_map = model(image)
edge_map = edge_map.squeeze().detach().numpy()
print(edge_map.shape)

threshold_value, _ = cv2.threshold(edge_map.astype(np.uint8), 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# apply thresholding to edge map
edge_map = (edge_map >= threshold_value) * 255

plt.imshow(edge_map.astype(np.uint8), cmap='gray')
cv2.imwrite('/content/drive/MyDrive/image/image_rgb_1/edge_1.jpg',edge_map)
plt.axis('off')
plt.show()
```

```python
img = cv2.imread('/content/drive/MyDrive/image/image_rgb_1/edge.jpg', 0) 

# Otsu's thresholding
otsu_thresh, otsu = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# hard thresholding
hard_thresh = 127
hard = cv2.threshold(img, hard_thresh, 255, cv2.THRESH_BINARY)[1]

# mean adaptive thresholding
mean_block_size = 11
mean_c = 2
mean = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, mean_block_size, mean_c)

# Gaussian adaptive thresholding
gaussian_block_size = 11
gaussian_c = 2
gaussian = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, gaussian_block_size, gaussian_c)

plt.imshow(otsu)
plt.title('Otsu thresholding')
plt.show()
plt.imshow(hard)
plt.title('Hard thresholding')
plt.show()
plt.imshow(mean)
plt.title('Mean adaptive thresholding')
plt.show()
plt.imshow(gaussian)
plt.title('Gaussian adaptive thresholding')
plt.show()
cv2.imwrite('/content/drive/MyDrive/image/image_rgb_1/otsu_thresholded.jpg', otsu)
cv2.imwrite('/content/drive/MyDrive/image/image_rgb_1/hard_thresholded.jpg', hard)
cv2.imwrite('/content/drive/MyDrive/image/image_rgb_1/mean_thresholded.jpg', mean)
cv2.imwrite('/content/drive/MyDrive/image/image_rgb_1/gaussian_thresholded.jpg', gaussian)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
